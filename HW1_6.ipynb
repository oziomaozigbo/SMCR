```python
# comments wi th the purpose of t h i s module and your name
import numpy as np
import matplotlib.pyplot as plt

N = int(input("enter fibonacci limit: "))
def seqnce(N):
    fib = np.empty(N)
    fib[0]=1
    fib[1]=1
    m = 2
#returns a Fibonacci sequence of any length N, N greater than 1.
    for m in range(2, N):
        fib[m] = fib[m-1] + fib[m-2]
        m+1
    return(fib)

#...code to
#(a)compute the Fibonacci sequence F_1 . . . F_N
F = seqnce(N)

#(b)compute the array of indices j = [1 . . . N] using np.linspace
#...about five lines of code
j = np.linspace(1,N,N)

#return j, F

print(F)
print(j)

#...code to
# (a) obtain the Fibonacci sequence F and the array j from the func t i on seqnce (N) ,

seqnce(N)
def plot(N):
  gold = np.empty(N)
  i = 0
  for i in range(0,N-1):
      gold[i] = F[i]/F[i+1]
  return(gold)

# (b) calculates the squence of golden ratio values from F
G = plot(N)

# (c) plots the Fibonacci sequenc and p l o t s the golden r a t i o sequence
# where the plot axes are labeled
#...about two lines of code for (a) and (b) plus another dozen for (c)

#plots fibonacci
L = len(F)
x = np.linspace(1,L,L)
y = F
plt.plot(x,y, color='blue', linestyle='--', marker='o', label='fibonacci plot')

plt.xlabel("Index")
plt.ylabel("Fibonacci number")
plt.title("Fibonacci Sequence")

#plots golden ratio
L = len(G)
x = np.linspace(1,L,L)
y = G
plt.plot(x,y, color='red', linestyle='--', marker='*', label='golden ratio plot')

plt.xlabel("Index")
plt.ylabel("golden ratio")
plt.title("Golden Ratio Sequence")



```
