```python
import numpy as np

m = 2 #number of rows of A
n = 5 #number of columns of A
A = np. random.randn(m, n) #an m by n matrix of random numbers
print(A)
AtA = A.T @ A #A‚àítranspose times A . . . n by n
AtA
AAt = A @ A.T #A‚àítranspose times A . . . n by n
AAt

sizeA = A.shape
rankA = np.linalg.matrix_rank(A)

sizeAtA = AtA.shape
rankAtA = np.linalg.matrix_rank(AtA)

sizeAAt = AAt.shape
rankAAt = np.linalg.matrix_rank(AAt)

print (" size (A) =", sizeA )
print (" rank (A) =", rankA )
print (" size ( AtA ) =", sizeAtA )
print (" rank ( AtA ) =", rankAtA )
print (" size ( AAt ) =", sizeAAt )
print (" rank ( AAt ) =", rankAAt )

m = 3 #number of rows of A
n = 4 #number of columns of A
A = np. random.randn(m, n) #an m by n matrix of random numbers
print(A)
AtA = A.T @ A #A‚àítranspose times A . . . n by n
AtA
AAt = A @ A.T #A‚àítranspose times A . . . n by n
AAt

sizeA = A.shape
rankA = np.linalg.matrix_rank(A)

sizeAtA = AtA.shape
rankAtA = np.linalg.matrix_rank(AtA)

sizeAAt = AAt.shape
rankAAt = np.linalg.matrix_rank(AAt)

print (" size (A) =", sizeA )
print (" rank (A) =", rankA )
print (" size ( AtA ) =", sizeAtA )
print (" rank ( AtA ) =", rankAtA )
print (" size ( AAt ) =", sizeAAt )
print (" rank ( AAt ) =", rankAAt )

m = 3 #number of rows of A
n = 1 #number of columns of A
A = np. random.randn(m, n) #an m by n matrix of random numbers
print(A)
AtA = A.T @ A #A‚àítranspose times A . . . n by n
AtA
AAt = A @ A.T #A‚àítranspose times A . . . n by n
AAt

sizeA = A.shape
rankA = np.linalg.matrix_rank(A)

sizeAtA = AtA.shape
rankAtA = np.linalg.matrix_rank(AtA)

sizeAAt = AAt.shape
rankAAt = np.linalg.matrix_rank(AAt)

print (" size (A) =", sizeA )
print (" rank (A) =", rankA )
print (" size ( AtA ) =", sizeAtA )
print (" rank ( AtA ) =", rankAtA )
print (" size ( AAt ) =", sizeAAt )
print (" rank ( AAt ) =", rankAAt )

#the rank values make sense because
#For an ùëö√óùëõ matrix the rank is equal to n if
#n < m and The n columns are linearly independent.

#For the case of m = 2
# what condition on A would result in the rank of AAT equaling 1?

m = 2 #number of rows of A
n = 5 #number of columns of A
A = np.random.randn(m, n) #an m by n matrix of random numbers
AAt = A @ A.T #A‚àítranspose times A . . . n by n
AAt
sizeAAt = AAt.shape
rankAAt = np.linalg.matrix_rank(AAt)

#For a m√óùëõ where m = 2, the rank of AAT is 1 if the two rows of are linearly dependent
#that is if one is a scalar multiple of the other
```
