```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import array

#without for loop
j= np.linspace(1,50,50)
print (j)
x = (4 / ((2*j)-1))
print (x)
y = ((-1)**(j+1))
print (y)
S = x@y
print (S)

#sums the function usin for loop

def sum_series(N):
    sum = 0
    SN=[]
    for j in range(1,51):
        sum = sum + (4/((2*j)-1))*((-1)**(j+1))
        SN.append(sum)
    return np.array(SN)

N = 50
SN = sum_series(N)

#Plot SN vs N
def plot(SN):
  plt.plot(range(1, N+1),SN, color='blue', linestyle='--', marker='o', label='Sumpi')

plot(SN)

plt.xlabel("N")
plt.ylabel("SN")
plt.title("SN vs N")
```
